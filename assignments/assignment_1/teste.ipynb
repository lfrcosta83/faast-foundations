```python


import pandas as pd
import numpy as np

def clean_data(fname_input, fname_output):
    df = pd.read_csv("../life_expectancy/data/" + fname_input, delimiter='\t')
    df[['unit', 'sex', 'age', 'region']] = df['unit,sex,age,geo\\time'].str.split(",", expand=True)
    df.drop(columns='unit,sex,age,geo\\time', inplace=True)
    df = pd.melt(df, id_vars=['unit', 'sex', 'age', 'region'], var_name="year")
    df = df.astype({'year':'int'})
    df.replace(": ", np.nan, inplace  = True)
    df['value'] = df['value'].str.split(' ').str[0]
    df = df.astype({'value':float}, errors="raise")
    df_pt = df[df["region"] == "PT"]
    df_pt.to_csv("../life_expectancy/data/" + fname_output, index=False, encoding='utf-8')
    return 

clean_data("eu_life_expectancy_raw.tsv", "pt_life_expectancy.csv")
```
